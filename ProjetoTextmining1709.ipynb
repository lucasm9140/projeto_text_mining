# 1 - Realizar a tokenização de palavras do artigo

```python
# Importa a biblioteca tokenize para a tokenização das palavras
from nltk.tokenize import word_tokenize
```

```python
# lista de tokens carregado do artigo
tokens = word_tokenize(artigo.cleaned_text)
```

```python
# Exibindo os tokens
print(tokens)
```

```python

import string # importa a (string) biblioteca, que fornece uma coleção de constantes e funções de string.
nltk.download('stopwords') # Esta linha baixa o stopwordscorpus da biblioteca Natural Language Toolkit (NLTK).
                          #Este corpus contém uma lista de palavras comuns (como "a", "the", "is", etc.)
                          #que são frequentemente removidas de dados de texto durante o pré-processamento.
from nltk.corpus import stopwords # Esta linha importa o stopwordsmódulo do nltk.corpuspacote. Este módulo fornece funções para acessar o corpus stopwords.
```

```python
def remove_punctuation(text): # função para remover a pontuação no texto
  no_punct = "".join([c for c in text if c not in string.punctuation])
  return no_punct
```

```python
tokens_no_punct = [remove_punctuation(token.lower()) for token in tokens]
```

```python
clean_tokens = [token for token in tokens_no_punct if token not in stopwords.words('portuguese')]
```

```python
print(clean_tokens)
```

```python
# Importando pandas e transformando em dataframe as palavras tokens do artigo
import pandas as pd
```

```python
df = pd.DataFrame(clean_tokens, columns=['Tokens'])
```

```python
display(df)
```

```python
# prompt: Usando o DataFrame df: grafico

# transformando o dataframe, token em grafico
import pandas as pd
import matplotlib.pyplot as plt

# Calculate token frequencies
token_counts = df['Tokens'].value_counts()

# Create a bar chart
plt.figure(figsize=(16, 6))
plt.bar(token_counts.index, token_counts.values)
plt.xlabel('Tokens')
plt.ylabel('Frequency')
plt.title('Token Frequency Distribution')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# 2 Realizar a distribuição de frequência das palavras (liberado geral o uso de novos stopwords) - FreqDist e Gráfico de linha.

```python
from nltk.probability import FreqDist #importa a FreqDistclasse do nltk.probability módulo.
                                      # FreqDist uma classe para calcular distribuições de frequência de palavras
                                      #ou outros itens em um texto.
import matplotlib.pyplot as plt #importa o pyplotmódulo da matplotlibbiblioteca e o atribui ao alias plt. matplotlib.pyploté uma coleção de funções que fazem matplotlib funcionar como MATLAB.
                                #Ela fornece uma interface conveniente para criar vários tipos de gráficos,
                                #incluindo gráficos de distribuição de frequência.
```

```python
fdist = FreqDist(tokens)
```

```python
fdist.plot()
```

```python
# Imprime as 10 palavras mais frequentes
print(fdist.most_common(10))
```

```python
# Plota a distribuição de frequência
fdist.plot(30, cumulative=False)
```

```python
plt.figure(figsize=(16, 8))  # Define o tamanho da figura
fdist.plot(30, cumulative=False, color='green')
plt.title("Distribuição de Frequência de Palavras")
plt.xlabel("Palavras")
plt.ylabel("Frequência")
plt.xticks(rotation=75)  # Rotaciona os labels do eixo x em 45 graus
plt.show()
```

```python
import requests
from bs4 import BeautifulSoup
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from nltk.probability import FreqDist
import nltk

# nltk.download('punkt')

# URL do artigo
url = "https://criancaenatureza.org.br/pt/clima/queimadas-desmatamento/"

# Faz a requisição HTTP para a URL
response = requests.get(url)

# Verifica se a requisição foi bem-sucedida
if response.status_code == 200:
    # Analisa o conteúdo HTML da página
    soup = BeautifulSoup(response.content, "html.parser")

    # Extrai o texto do artigo (ajuste o seletor CSS de acordo com a estrutura do site)
    # Use a more specific selector to target the article content
    article_content = soup.select_one('.entry-content')
    if article_content:
        texto = article_content.get_text()

        # Tokeniza o texto
        tokens = nltk.word_tokenize(texto)

        # Remove stop words (opcional)
        # nltk.download('stopwords')
        # stop_words = set(nltk.corpus.stopwords.words('portuguese'))  # Ou outra língua
        # tokens = [w for w in tokens if not w.lower() in stop_words]

        # Calcula a distribuição de frequência
        fdist = FreqDist(tokens)

        # Gera a wordcloud
        wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(fdist)

        # Exibe a wordcloud
        plt.figure(figsize=(10, 5))
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis("off")
        plt.show()
    else:
        print("Artigo não encontrado na página.")
else:
    print("Erro ao acessar a URL.")
```
